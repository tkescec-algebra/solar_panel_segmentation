```python
import cv2
import numpy as np
import os

# Putanja do mape s maskama
directory_path = '../data/val/masks'

# Petlja za svaku datoteku u direktoriju
for filename in os.listdir(directory_path):
    if filename.endswith(".png"):  # Provjerava da li je datoteka PNG format
        file_path = os.path.join(directory_path, filename)
        mask = cv2.imread(file_path, 0)  # Učitava sliku u grayscale modu

        # Dobivanje jedinstvenih vrijednosti
        unique_values = np.unique(mask)
        print(f"Jedinstvene vrijednosti u maski {filename}:", unique_values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_training_losses(train_losses):
    """
    Plots training losses for multiple models and loss functions.
    
    Args:
    train_losses (list of dicts): List where each dict contains model and loss function names as keys and list of losses as values.
    """
    plt.figure(figsize=(10, 6))
    for model_loss_dict in train_losses:
        for model_loss_name, losses in model_loss_dict.items():
            epochs = np.arange(1, len(losses) + 1)
            plt.plot(epochs, losses, label=model_loss_name)
    
    plt.title('Training Losses Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(title='Model and Loss Function', loc='upper right')
    plt.grid(True)
    plt.show()

# Example usage:
train_losses = [
    {'deeplabv3_resnet50_BCEWithLogitsLoss': [0.5, 0.4, 0.3, 0.2]},
    {'deeplabv3_resnet50_L1Loss': [0.6, 0.5, 0.4, 0.3]},
    {'fcn_resnet50_BCEWithLogitsLoss': [0.45, 0.35, 0.25, 0.15]},
    {'fcn_resnet50_L1Loss': [0.55, 0.45, 0.35, 0.25]},
    {'custom_unet_BCEWithLogitsLoss': [0.65, 0.55, 0.45, 0.35]},
    {'custom_unet_L1Loss': [0.75, 0.65, 0.55, 0.45]}
]

plot_training_losses(train_losses)
```
